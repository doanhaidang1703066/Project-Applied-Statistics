```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc dữ liệu từ file CSV đã lưu
csv_path = "/Users/doanhaidang1703/Documents/Applied Statistics/Project Applied Stats/code/results/evaluation_metrics.csv"
try:
    df = pd.read_csv(csv_path)
except FileNotFoundError:
    print(f"Không tìm thấy file {csv_path}. ")
    raise

# 2. Thiết lập giao diện tổng thể
sns.set_theme(style="whitegrid")
metrics = ['Accuracy', 'Precision', 'Recall', 'F1_Score', 'Time_sec']
datasets = df['Dataset'].unique()

# Cài đặt bảng màu chuyên nghiệp
colors = sns.color_palette("Set2")

# 3. Vòng lặp vẽ 1 Dashboard (5 biểu đồ) cho từng Dataset
for dataset in datasets:
    # Lọc dữ liệu riêng cho dataset hiện tại
    df_subset = df[df['Dataset'] == dataset]
    
    # Tạo một Figure lớn chứa 5 biểu đồ con (1 hàng, 5 cột)
    fig, axes = plt.subplots(1, 5, figsize=(22, 5))
    fig.suptitle(f"Báo Cáo Hiệu Suất: Tập dữ liệu {dataset.upper()}", fontsize=18, fontweight='bold', y=1.05)
    
    for i, metric in enumerate(metrics):
        ax = axes[i]
        
        # Vẽ biểu đồ cột
        sns.barplot(data=df_subset, x='Model', y=metric, ax=ax, palette=colors, hue='Model', legend=False)
        
        # Trang trí tiêu đề và trục
        ax.set_title(metric, fontsize=14, fontweight='bold')
        ax.set_xlabel('')
        ax.tick_params(axis='x', rotation=45) # Xoay tên model cho đỡ bị đè chữ
        
        # Gắn label giá trị trực tiếp lên đầu mỗi cột
        for p in ax.patches:
            val = p.get_height()
            ax.annotate(f"{val:.3f}", 
                        (p.get_x() + p.get_width() / 2., val), 
                        ha='center', va='bottom', 
                        xytext=(0, 5), textcoords='offset points',
                        fontsize=10)
        
        # Xử lý riêng trục Y
        if metric == 'Time_sec':
            ax.set_ylabel('Thời gian (Giây)')
            # Nếu thời gian chênh lệch quá lớn (VD: k-NN 0.01s vs LSTM 5s), có thể bật log-scale:
            # ax.set_yscale('log') 
        else:
            ax.set_ylim(0, 1.1) # Cố định thang đo 0-1 cho Acc, F1, Recall, Prec
            ax.set_ylabel('Điểm số')

    # Chống dính các biểu đồ vào nhau
    plt.tight_layout()
    plt.show()
    print("-" * 100) # Đường viền ngăn cách giữa các dataset
```
